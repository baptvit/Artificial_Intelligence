```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def distancia(a1, b1, a2, b2):
    return np.sqrt((a1-a2)**2+(b1-b2)**2)

qtdeLinhas = 200
qtdeColunas = 200


W = np.zeros((qtdeLinhas,qtdeColunas))

centro = [100,100]
raio = 50

for i in range(qtdeLinhas):
    for j in range(qtdeColunas):
        if(distancia(i,j,centro[0],centro[1]) < raio):
            W[i,j] = 255

    
plt.gray()
plt.imshow(W)


```

Dado um pixel p(i,j), quais são seus vizinhos?

![alt text](vizinhanca.jpg "Vizinhos de um pixel")

Vizinhança-de-4: A = p(i-1,j), B = p(i, j+1), C = p(i+1, j), D = p(i, j-1)

Vizinhança-de-8: A = p(i-1,j-1), B = p(i-1, j), C = p(i-1, j+1), D = p(i, j+1), E = p(i+i, j+1), F = p(i+1, j), G = p(i+1, j-1), H = p(i,j-1)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bolinhas.png')

qtdeLinhas, qtdeColunas, c = img.shape
W = np.zeros((qtdeLinhas,qtdeColunas))

B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]

Gray = B/3 + G/3 + R/3

for i in range(1, qtdeLinhas-1):
    for j in range(1, qtdeColunas-1):
       
        A = Gray[i-1,j]
        B = Gray[i, j+1]
        C = Gray[i+1, j]
        D = Gray[i, j-1]
        
        P = Gray[i,j]
        
        if(A > P or B > P or C > P or D > P):
            W[i,j] = 255

  
plt.gray()
plt.imshow(W)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bolinhas.png')

qtdeLinhas, qtdeColunas, c = img.shape
W = np.zeros((qtdeLinhas,qtdeColunas))

B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]
  

for i in range(qtdeLinhas):
    for j in range(qtdeColunas):
        if(B[i,j] == 255 and G[i,j] == 255 and R[i,j] == 255):
            W[i,j] = -1
      
cont = 0


for i in range(1, qtdeLinhas-6):
    for j in range(1, qtdeColunas-6):
        if(W[i,j] == 0):
            A = W[i-1,j-1]
            B = W[i-1,j]
            C = W[i-1,j+1]
            D = W[i,j-1]
            E = W[i-1,j+2]
            F = W[i-1,j+3]
            G = W[i-1,j+4]
            H = W[i-1,j+5]
            if(A == -1 and B == -1 and C == -1 and D == -1 and E == -1 and F == -1 and G == -1 and H == -1):
                cont = cont + 1
                W[i,j] = cont
            else:
                W[i,j] = max(A,B,C,D,E,F,G,H)
    
    
    
    
plt.gray()
plt.imshow(W)
```
