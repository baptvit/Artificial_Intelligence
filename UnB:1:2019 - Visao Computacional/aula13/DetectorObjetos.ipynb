```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt


detectorObjeto = cv.CascadeClassifier('detector_5.xml')
img = cv.imread('teste.bmp')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

objetos_detectados = detectorObjeto.detectMultiScale(gray, 1.01, 1)
for (x,y,w,h) in objetos_detectados:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        
plt.imshow(img)
```
